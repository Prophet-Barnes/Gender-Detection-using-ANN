### Splitting the DataFrame on the basis of full values and NaN values in '_last_judgement_at':

```python
last_judgement_at_empty = df1[df1['_last_judgment_at'].isna()]
last_judgement_at_empty
```

### Filling empty values with 'unknown' :

```python
last_judgement_at_empty['_last_judgment_at'].fillna(value = 'unknown', inplace = True)
last_judgement_at_empty
```

```python
last_judgement_at_not_empty = df1[df1['_last_judgment_at'].notna()]
last_judgement_at_not_empty
```

### Changing the format of the column '_last_judgment_at' by splitting the values to retain only the date and not the time

```python
df1['_last_judgment_at'].value_counts()
```

```python
for i in range(last_judgement_at_not_empty.shape[0]):
    current_value = last_judgement_at_not_empty['_last_judgment_at'].values[i]
    list_of_values = current_value.split(' ')
    last_judgement_at_not_empty['_last_judgment_at'].values[i] = list_of_values[0]
```

```python
df1 = last_judgement_at_not_empty.append(last_judgement_at_empty)
df1
```

# Preparing the Data to train Our Neural Network

## Balancing the Dataset :

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler()
```

```python
X = df1[['_golden', '_unit_state', '_trusted_judgments', '_last_judgment_at',
       'gender:confidence', 'profile_yn:confidence', 'created',
       'fav_number', 'gender_gold', 'link_color', 'profile_yn_gold',
       'retweet_count', 'sidebar_color', 'tweet_count', 'user_timezone']]
Y = df1[['gender']]
```

```python
len(X.columns)
```

```python
X_bal, Y_bal = rus.fit_resample(X,Y)
```

```python
Y_bal['gender'].value_counts()
```

## Normalizing the Dataset :

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = MinMaxScaler().fit_transform(X_bal.values)
Y = Y_bal.values
```

## Splitting the Dataset for training and testing :

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, random_state = 0)
```

## Optional : Importing RandomForestClassifier and SVC for comparing accuracy of our Neural Network with those of classical Machine Learning Models

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.svm import SVC
```

```python
rfc = RandomForestClassifier()
svm = SVC()
```

```python
svm.fit(X_train, Y_train)
rfc.fit(X_train, Y_train)
```

## Importing accuracy_score :

```python
from sklearn.metrics import accuracy_score
```

```python
Y_pred_rfc = rfc.predict(X_test)
Y_pred_svm = svm.predict(X_test)
```

```python
accuracy_score(Y_test, Y_pred_svm)
```

```python
accuracy_score(Y_test, Y_pred_rfc)
```

## 2) Using TensorFlow :

```python
from tensorflow import keras
```

```python
from tensorflow.keras.layers import Input, Dense, Activation
```

```python
from tensorflow.keras import Sequential
```

```python
from keras.callbacks import ModelCheckpoint
```

```python
tfnn = Sequential()
```

#### Note : The accuracy of the Neural Network didn't change at all for any of the tried combinations for the sizes of hidden layers (not even by the slightest mark). Hence, the simplest Neural Network is opted in this program for performance with respect to time.

```python
tfnn.add(Input(shape = (15,)),)
tfnn.add(Dense(25, activation = 'relu'),)
tfnn.add(Dense(20, activation = 'relu'),)
tfnn.add(Dense(15, activation = 'relu'),)
tfnn.add(Dense(4, activation = 'softmax'),)
```

```python
tfnn.compile(optimizer='adam', loss='SparseCategoricalCrossentropy', metrics = ['accuracy'])
```

```python
cp = ModelCheckpoint('best_one', verbose=0, save_best_only=True)
```

```python
tfnn.fit(x = X_train, y = Y_train, epochs=30, callbacks=[cp], validation_split=0.2)
```

```python
from keras.models import load_model
tfnn = load_model('best_one')
```

```python
tfnn.predict(X_test)
```
